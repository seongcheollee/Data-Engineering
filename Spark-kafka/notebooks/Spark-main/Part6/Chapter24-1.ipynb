#고급 분석과 머신러닝 개요
- 스파크 고급 분석 도구
  - 데이터 전처리: 데이터 정제 및 피처 엔지니어링
  - 지도 학습
  - 비지도 학습
  - 추천 엔진
  - 그래프 분석
  - 딥러닝

## 고급 분석에 대한 짧은 입문서
- 고급 분석이란, 데이터 기반의 인사이트를 도출하여 핵심 문제를 해결하거나 예측 또는 추천을 하기 위한 기술을 의미
- 머신러닝에서 가장 일반적으로 활용되는 작업
  - 다양한 특징(feature)을 기반으로 각 샘플에 부여된 레이블을 예측하는 분류/회귀 문제를 포함한 지도학습
  - 사용자의 과거 행동에 기반하여 제품을 제안하는 추천엔진
  - 군집 분석, 이상징후 탐지, 토픽 모델링 등 데이터 구조를 파악하기 위한 비지도 학습
  - 소셜 네트워크상에서 유의한 패턴을 찾는 그래프 분석

### 지도 학습
- 레이블(종속변수)을 포함하는 과거 데이터를 사용하여 모델을 학습하고 데이터의 다양한 특징을 기반으로 레이블값을 예측하는 행위
  - 이러한 학습은 경사하강법(gradient descent)과 같은 반복적 최적화 알고리즘으로 진행됨
  - 학습 알고리즘은 기본 모델로 시작하여 반복적으로 진행되는 학습 과정에서 다양한 내부 파라미터를 조정하며 단계적으로 모델 개선
  - ex) 연령(특징)을 기반으로 개인의 소득(종속변수)을 예측
  
- 지도학습은 예측하고자 하는 변수의 타입에 따라 분류와 회귀로 구분함

#### 분류
- 범주형(불연속적이고 유한한 값의 집합) 종속변수를 예측하는 알고리즘을 학습시키는 행위
  - 이진 분류: 두 그룹 중 하나로 분류
    - ex)스팸이냐 아니냐
  - 다중 클래스 분류: 세 가지 이상의 범주로 분류
    - ex)이메일을 4가지 범주로 구분 -> 스팸/개인/업무/기타
    

- 분류 사례
  - 질병 예측
    - 과거 환자의 행동이나 생리학적 특성에 대한 데이터셋을 학습하여 환자가 심장 질환을 앓고 있는지 아닌지 예측
    - 이진 분류: 건강한 심장 / 건강하지 않은 심장
  - 이미지 분류
    - 애플, 구글, 페북 등의 인터넷 기업들은 과거 사진첩에 담긴 사람들의 이미지를 학습에 활용할 수 있음
    - 주어진 사진의 인물이 누구인지 맞추는 문제 등에 적용 가능
  - 고객 이탈 예측
    - 과거 이탈했거나 유지했던 이력이 담긴 고객 데이터를 학습하여 미래 이탈여부를 예측
  - 구매 여부 예측
    - 사용자의 인터넷 검색 패턴, 위치 관련 데이터를 학습하여 웹사이트 방문자의 특정 제품 구매여부를 예측

#### 회귀
- 연속형 변수(실수)를 예측하는 알고리즘을 학습시키는 행위
- 회귀 사례
  - 판매량 예측
    - 과거 판매 데이터를 분석하여 총 제품 판매량 예측
    - 지난주의 판매 데이터를 이용하여 다음날 판매량 예측 가능
  - 신장(키) 예측
    - 부모의 신장을 활용하여 미래 자녀의 신장 예측
  - 관객 수 예측
    - 넷플같은 미디어 기업은 얼마나 많은 가입자가 특정 콘텐츠를 볼 것인지 예측하려고 함

### 추천
- 고급분석에서 가장 직관적인 응용 영역 중 하나
  - 다양한 제품이나 명시적인 선호도(등급 등)나 암시적인 선호도(관찰된 행동을 통해 도출)를 연구하여 사용자 또는 아이템 간의 유사성을 도출
  - 유사성을 기반으로 특정 사용자와 유사한 사용자가 선호하는 상품을 추천하거나 이미 구매한 상품과 비슷한 상품을 추천
- 추천 사례
  - 영화 추천
    - 넷플은 스파크를 사용한 대규모 영화 추천 서비스를 제공함
      - 사용자가 어떤 영화를 보았고, 어떤 영화를 보지 않았는지 연구하고
      - 해당 영화에 대한 사용자의 평점이 다른 사용자와 얼마나 유사한지 반영함
  - 상품 추천
    - 아마존에서 상품 추천은 매출 증대를 위한 핵심 도구임
      - 고객에게 상품을 추천하기 위해 고객의 장바구니를 구성하는 아이템을 기반으로 과거 장바구니 이력을 탐색하고,
      - 가장 유사한 아이템을 보유한 장바구니에 포함된 다른 아이템을 추천함
      - 또한 각 상품 페이지마다 다른 고객이 구매한 유사 상품을 보여줌

### 비지도 학습
- 주어진 데이터셋에서 특정 패턴을 찾거나 숨겨진 구조적 특징을 발견하는 행위
- 예측 대상이 되는 종속변수(레이블)가 필요없음
- 지도 학습에 비해 좋고 나쁨을 결정하기가 비교적 어려움
- 사례
  - 이상징후 탐지
    - 시간의 흐름에 따라 주기적으로 발생하는 표준 이벤트 외 비표준 이벤트 발생하는 경우 탐지
      - 특정 구역의 보안 관리자는 사람만 다닐 수 있는 통로에 갑자기 자동차, 자전거 등이 나타났을 때 알림을 받고 싶어할 것임
  - 사용자 세분화
    - 특정 사용자가 다른 사용자와 어떤 속성이 유사한지 더 잘 이해할 수 있음
      - 게임 회사는 게임 이용 시간이라는 속성을 기반으로 사용자를 군집화함
        - 캐주얼한 플레이어와 하드코어 플레이어가 매우 다른 행동을 한다는 것을 파악하여 각 플레이어에게 다른 추천이나 보상을 제공    
  - 토픽 모델링
    - 문서의 다양한 단어를 분석하여 단어 간의 내재된 연관성이 있는지 살펴볼 수 있음
      - 웹페이지들을 각 주제로 군집화함

### 그래프 분석 
- 객체를 가리키는 정점(vertex)과 객체 간의 관계를 나타내는 에지(edge)를 지정하는 구조에 대한 연구
- 그래프는 관계에 관한 모든 것을 의미하므로 관계를 지정하는 모든 것이 그래프 분석의 사례가 될 수 있음
- 사례
  - 사기거래 예측
    - 미국의 금융지주회사인 캐피탈원은 스파크의 그래프 분석 기능을 사기 네트워크를 감지하는데 사용했음
      - 과거 사기 이력이 있는 사람의 정보(폰번호, 주소, 이름 등)를 활용하여 의심스러운 여신 요청이나 거래 탐지
      - 네트워크상의 인접성 정보를 활용하여, 사기성 전화번호와 가까운 거리에 존재하는 모든 사용자 계정을 의심
  - 이상징후 탐지
    - 각 개체의 네트워크가 서로 어떻게 연결되어 있는지 살펴봄으로써 담당자가 추가로 조사할 특잇값 및 이상값을 식별
    - ex) 대부분의 각 정점이 연관된 10개의 에지와 연결되어 있는데 특정 정점은 1개의 에지와 연결되어 있음
  - 네트워크 특성 분류
    - 네트워크 내의 특정 정점에 관한 몇 가지 특성이 주어지면, 이미 존재하는 노드와의 연결구조에 따라 다른 정점을 분류할 수 있음
    - ex) 어떤 사람이 소셜 네트워크 내에서 \*영향력자로 식별되었을 때, 그 사람과 유사한 네트워크 구조를 보이는 다른 사람 역시 영향력자로 분류할 수 있음
  - 웹페이지 추천
    - 구글의 독창적인 웹 추천 알고리즘인 페이지랭크는 특정 웹페이지의 중요성을 평가하기 위해 웹사이트의 관계를 분석하는 그래프 알고리즘
  
- 네트워크 이론
  - \*영향력자: 네트워크 내에서 연결성이 강하고 타 정점에 영향을 줄 수 있는 정점
  - 사회적 영향력: 영향력자가 실제로 타 정점의 특성에 영향을 주는 현상
  - 구조적 등위성: 한 정점이 다른 정점과 직접적인 관계는 없지만, 동일한 네트워크 연결 형상을 나타내는 경우
    - 구조적 등위성을 가지는 두 정점은 동일한 역할을 수행할 수 있고, 서로 대체될 수 있는 관계임

### 고급 분석 프로세스
1. 분석 주제와 관련된 데이터 수집
2. 데이터 파악을 위해 데이터를 정제하고 검토함
3. 알고리즘이 데이터를 잘 인식하고 활용할 수 있도록 피처 엔지니어링을 수행
4. 후보 모델을 생성하는 하나 이상의 알고리즘을 학습시키기 위해 전체 데이터의 일부를 학습 데이터셋으로 사용
5. 나머지 데이터셋을 모델에 적용하여 결과를 객관적으로 측정하고, 분석 수행 시 정해 놓은 성공 기준과 비교하여 해당 모델을 최종 평가함
6. 앞의 과정을 거치며 확보한 통찰력 또는 모델을 사용하여 예측을 수행하거나 이상 현상을 감지하여 비즈니스 문제 해결

#### 데이터 수집 
- 일반적으로 데이터 수집의 목적은 알고리즘을 학습시키는 데 활용하고자 모으는 것

#### 데이터 정제 
- 데이터를 정제하고 검토하는 과정을 EDA(Exploratory Data Analysis)
- 일반적으로 대화형 쿼리 및 시각화 방법 등을 통해 데이터의 분포, 상관관계, 기타 세부 현황을 파악하는 과정임
- 오류를 방지하기 위해서는 분석하고자 하는 데이터를 사전에 명확히 파악해야함

#### 피처 엔지니어링 
- 머신러닝 알고리즘에 적용 가능한 형식으로 데이터를 변환하는 과정
  - 데이터 정규화
  - 파생 변수 생성
  - 범주형 변수 조작 등
- MLlib에서는 실젯값의 형태와 관계없이 모든 변수가 실수형 벡터로 입력되어야함

#### 모델 학습 
- 모델 학습 단계는 입력을 받았을 때 그에 적합한 출력을 예측하는 모델을 학습시키는 단계임
- 이 단계에서는 모델이 입력 데이터에 대해 얼마나 잘 동작하는지에 따라 모델 내부 파라미터가 달라질 수 있음
  - ex) 스팸메일 이진분류기 학습
    - 알고리즘이 특정 단어가 다른 단어보다 스팸메일을 더 잘 예측하는 것을 발견 -> 해당 단어와 연관된 파라미터에 더 높은 가중치 부여
- 이러한 학습 과정을 모델이라하며, 통찰력을 얻거나 미래를 예측하는 데 활용

d #### 모델 튜닝 및 평가 
 - 데이터를 여러 개로 나누는 것은, 새로운 데이터에 대해서도 일반화 가능한 모델을 생성해야하기 때문에 아주 중요한 과정임
   - 학습셋 / 검증셋 / 테스트셋
 - 검증셋으로는 학습에 영향을 주는 다양한 하이퍼파라미터를 시험해보고 모델의 적합성을 테스트할 수 있음
 - 테스트셋으로는 새로운 데이터에 일반화되지 않는 과적합 현상이 발생하는지 여부를 확인할 수 있음

#### 모델 및 통찰력 활용하기
- 학습 및 테스트 과정을 통해 성능이 좋은 모델이 만들어지면 비로소 모델을 사용할 수 있음

## 스파크의 고급 분석 툴킷
- 스파크에는 고급 분석을 수행하기 위한 몇 가지 핵심 패키지와 다양한 외부 패키지가 포함되어 있음
- 그 중에서 기본이 되는 패키지는 MLlib

### MLlib이란
- 제공하는 기능
  - 데이터 수집
  - 데이터 정제
  - 특징 추출 및 선택
  - 대규모 데이터를 대상으로한 지도 및 비지도 학습 ML모델의 학습과 튜닝
  - 모델을 운영환경에서 사용할 수 있도록 도와주는 인터페이스 
  
- 비고
  - MLlib는 서로 다른 핵심 데이터 구조를 활용하는 두 개의 패키지로 구성됨
  - pyspark.ml
    - DataFrame 전용
  - pyspark.mllib
    - 저수준 RDD 전용

#### MLlib는 언제 그리고 왜 사용해야할까
- 단일 머신 기반의 도구(사이킷런, 텐서플로, R패키지 등)와 MLlib는 <strong>상호보완적</strong>이라고 할 수 있음
  - <strong>확장성 문제</strong>에 마주하게 되는 경우 스파크로 해결 가능
- 스파크의 확장 기능이 적용되는 두 가지 주요 활용 사례
  - 스파크로 <strong>대량의 데이터를 전처리</strong> 후 단일 머신 기반의 학습 라이브러리 활용
  - <strong>입력 데이터나 모델 크기</strong>가 단일 머신에 올려놓기 어려운 경우 스파크로 분산 처리 기반 머신러닝 수행
    - 단, 학습 및 데이터 처리는 간단해지지만 모델 배포할 땐 최적화 기능 제공 X

## 고수준 MLlib의 개념
- MLlib에는 몇 가지 기본적인 구조적 유형이 있음
  - 변환자
  - 추정자
  - 평가기(evaluator)
  - 파이프라인
- 변환자
  - 원시 데이터를 다양한 방식으로 변환하는 함수
    - 새로운 상호작용 변수 생성
    - 컬럼 정규화
    - 타입 변경 등
  
- 추정자
  - 사용자가 데이터로부터 모델을 학습시키기 위해 사용하는 알고리즘
    - 알고리즘별로 머신러닝 개발단계를 추상화 해놓은 객체
    - 사용자가 직접 알고리즘을 개발하지 않아도 이미 구현된 모델을 쉽게 가져다 사용할 수 있는 고수준 API
    
- 평가기
  - 모델의 성능이 지정한 기준에 따라 어떻게 작동하는지 볼 수 있게함
  
- 파이프라인
  - 머신러닝 워크플로를 형성하기 위해 <strong>특정 순서로 다수의 변환자, 추정자를 연결함</strong>
  - 고수준에서 차례로 변환 -> 추정 -> 평가를 하나씩 지정할 수 있지만, 단계로 지정하는 것이 더 간편할 수 있음

### 저수준 데이터 타입
- 머신러닝 모델에 전달되는 일련의 특징은 <strong>Double 타입으로 구성된 벡터 형태</strong>로 전달되어야함
- 벡터는 <strong>희소(대부분의 요소가 0)</strong>하거나 <strong>밀도가 높을 수(고윳값이 많은 경우)</strong> 있음

```
from pyspark.ml.linalg import Vectors
```

#### Dense Vector
- 모든 값의 배열을 지정할 수 있음

```
denseVec = Vectors.dense(1.0,2.0,3.0)
denseVec
```

#### Sparse Vector
- 희소한 벡터를 만들기위해서는 벡터의 총 크기와 0이 아닌 요소의 인덱스 및 값을 지정할 수 있음
  - 압축하여 표현 가능

```
size = 3 #벡터의 총 사이즈
idx = [1,2] #벡터 내 0이 아닌 원소의 위치
values = denseVec[idx]
values
```

```
sparseVec =  Vectors.sparse(size, idx, values)
sparseVec
```
